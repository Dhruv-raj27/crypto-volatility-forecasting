```python
# notebooks/EDA_Bitcoin.ipynb

import pandas as pd
import matplotlib.pyplot as plt

# Load the processed data
df = pd.read_csv('../outputs/processed_Bitcoin.csv', parse_dates=['Date'], index_col='Date')

# Plot 1: Close Price
plt.figure(figsize=(12, 4))
plt.plot(df['Close'], label='Close Price')
plt.title('Bitcoin Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot 2: Log Returns
plt.figure(figsize=(12, 4))
plt.plot(df['Log_Return'], label='Log Return', color='orange')
plt.title('Bitcoin Daily Log Returns')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot 3: Rolling Volatility (7-day)
plt.figure(figsize=(12, 4))
plt.plot(df['Volatility_7d'], label='7-day Rolling Volatility', color='green')
plt.title('Bitcoin 7-Day Realized Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility (Standard Deviation)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
